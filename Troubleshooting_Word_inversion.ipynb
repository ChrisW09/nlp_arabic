# Notebook to read in PDF with content in arabic language

```python
# please uncomment to install package
# pip install pdfplumber

import pdfplumber
```

```python
# specify file path
file_path = 'read_pdfs/the-islamic-state-al-nabacc84_-newsletter-1.pdf'
```

```python
# read PDF pagewise
with pdfplumber.open(file_path) as stream_pdf:
    pages_content = [i.extract_text() for i in stream_pdf.pages]
```

```python
# nice printout
[print(i) for i in pages_content]
```

## Focus on Lamedh, Alpeh, Nun

```python
# decode content as binary
pages_content_binary = [i.encode('utf-8') for i in pages_content]
```

```python
#print to compare page 12 (short one)
print(pages_content_binary[11])
print('\n')
print(pages_content[11])
```
